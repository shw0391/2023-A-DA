```python
import numpy as np
import pandas as pd

index = pd.date_range('12/1/2024',periods=8)
df=pd.DataFrame(np.random.rand(8,3),index=index,columns=list('ABC'))
df
```

```python
df['D']=df['A']/df['B']
df
```

```python
df['E']=np.sum(df,axis=1)
df
```

```python
df['F']=np.ones(8)
df.head(3)
```

```python
df1=df.drop(['E','F'],axis='columns')
df1.head()
```

```python
del df1['D']
df1.head()
```

```python
df1.rename(columns={'a':'2020','b':2021},inplace=True)
df1.tail()
```
