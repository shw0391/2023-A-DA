## plt.figure(figsize=(m,n))

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

f1=plt.figure(figsize=(10,2))
plt.plot(np.random.randn(100))

plt.title('figure size:(10,2)')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(5,3))
plt.subplot(2,3,1)
plt.subplot(2,3,5)
plt.subplot(3,5,6)
plt.show()
```

```python
ax1 = plt.subplot(2,1,1)
plt.plot(np.random.randint(1,10,10),'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')
print(ax1)

ax2 = plt.subplot(2,1,2)
plt.plot(np.arange(10),np.arange(10)*2,'r.-')
plt.xlabel('time(s)')
plt.ylabel('Undamped')
print(ax2)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

fig,axes = plt.subplots(2, 2)
plt.figure(figsize=(6,4))

axes[0,0].plot([1,2,3,4])
axes[0,0].set_title('subplot1')

axes[0,1].plot([4,3,2,1])
axes[0,1].set_title('subplot2')

axes[1,0].plot([1,2,2,3,3,3,4,4,4,4])
axes[1,0].set_title('subplot3')

axes[1,1].plot([1,2,3,4],[4,3,2,1])
axes[1,1].set_title('subplot4')

plt.tight_layout()
plt.show()
```
